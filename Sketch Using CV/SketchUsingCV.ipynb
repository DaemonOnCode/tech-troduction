```
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
```

```
img = cv2.imread("SRMKZILLA Logo.jpg")
img.shape
img = cv2.resize(img , (int(img.shape[1]*1), int(img.shape[0]*1)),interpolation=cv2.INTER_AREA)
```

```
cv2_imshow(img)
print('Coloured')
cv2.waitKey()
cv2.destroyAllWindows()
```

```
gray = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)

cv2_imshow(gray)
print('Black and White')
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```
inverse = 255-gray
gauss = cv2.GaussianBlur(inverse,ksize=(15,15), sigmaX =0 , sigmaY = 0)

sketch = cv2.divide(gray , 255-gauss , scale=255)

cv2_imshow(sketch)
print('Sketch')
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```
cv2_imshow(img)
print('Coloured')
cv2_imshow(gray)
print('Black and White')
cv2_imshow(sketch)
print('Sketch')
cv2.waitKey(0)
cv2.destroyAllWindows()
```
